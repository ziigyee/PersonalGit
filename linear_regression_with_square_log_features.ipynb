```python
import numpy as np

import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline 
plt.style.use('fivethirtyeight')
```

```python
np.random.seed(42)

population = np.linspace(1000,100000,50)
```

```python
#electricity = np.add(10**10 + population**4,np.random.normal(0,10*40,size=len(population)))
```

```python
electricity = np.add(10**10 + population**2,np.random.normal(0,0.3*(10**9),size=len(population)))
```

```python
plt.scatter(population,electricity)
plt.title('Population agains electricity')
plt.xlabel('population')
plt.ylabel('electricity')
plt.show()
```

### Prediction

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(population.reshape(-1, 1),electricity)
predictions = lr.predict(population.reshape(-1, 1))
```

```python
plt.scatter(population,electricity,label='real',s=10)
plt.scatter(population,predictions,label='predicted',s=10)
plt.legend()
plt.title('Population agains real electricity and predicted electricity')
plt.xlabel('population')
plt.ylabel('electricity')
plt.show()
```

### Let's trasform the feature

```python
new_x = population**2
```

```python
plt.scatter(new_x,electricity)
plt.title('Population**2 vs. Electricity')
plt.xlabel('population**2')
plt.ylabel('electricity')
```

```python
# The formula now is electricity = b0 + b1 * population**2

lr = LinearRegression()
lr.fit(new_x.reshape(-1, 1),electricity)
predictions = lr.predict(new_x.reshape(-1, 1))
```

```python
plt.scatter(population,electricity,label='real',s=10)
plt.scatter(population,predictions,label='predicted',s=10)
plt.legend()
plt.title('Population agains real electricity and predicted electricity')
plt.xlabel('population')
plt.ylabel('electricity')
plt.show()
```

# Log feature

```python
age = np.random.randint(20,80,size=(400))
```

```python
salary = np.add(np.log2(age)*10000, np.random.normal(0,1000,size=len(age)))
```

```python
plt.scatter(age,salary)
plt.title('Age agains Salary')
plt.xlabel('age')
plt.ylabel('salary')
plt.show()
```

```python


lr = LinearRegression()
lr.fit(age.reshape(-1, 1),salary)
predictions = lr.predict(age.reshape(-1, 1))
```

```python
plt.scatter(age,salary,label='real',s=10)
plt.scatter(age,predictions,label='predicted',s=10)
plt.legend()
plt.title('Age agains real salary and predicted salary')
plt.xlabel('age')
plt.ylabel('salary')
plt.show()
```

### Let's tranform the feature

```python
new_age = np.log(age)
```

```python
plt.scatter(new_age,salary)
plt.title('log(Age) vs. Salary')
plt.xlabel('log(age)')
plt.ylabel('Salary')
```

```python
# The formula now is salary = b0 + b1*log(age)

lr = LinearRegression()
lr.fit(new_age.reshape(-1, 1),salary)
predictions = lr.predict(new_age.reshape(-1, 1))
```

```python
plt.scatter(age,salary,label='real',s=10)
plt.scatter(age,predictions,label='predicted',s=10)
plt.legend()
plt.title('Age agains real salary and predicted salary')
plt.xlabel('age')
plt.ylabel('salary')
plt.show()
```
