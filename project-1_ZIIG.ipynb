## 1. Defining a player

---

The player variables are:

    player_id : id code unique to each player (integer)
    player_name : entered name of the player (string)
    time_played : number of time played the game in minutes (float)
    player_pokemon: the player's captured pokemon (dictionary)
    gyms_visited: ids of the gyms that a player has visited (list)
    
Create the components for a player object by defining each of these variables. The dictionary and list variables should just be defined as empty; you can use any (correctly typed) values for the others.

````python
#Create a player object to store the attributes.
class Player:

    def __init__(self, player_id, player_name, time_played, player_pokemon, gyms_visited):
        
        #Check for Valid entries
        if type(player_id) is not int: raise ValueError, "Please Enter an Integer for player_id."
        if type(player_name) is not str: raise ValueError, "Please Enter a String for player_name."
        if type(time_played) is not float: raise ValueError, "Please Enter a Float for time_played."
        if type(player_pokemon) is not dict: raise ValueError, "Please Enter a Dictionary for player_pokemon."
        if type(gyms_visited) is not list: raise ValueError, "Please Enter a List for gyms_visited."
        
        #Alternative Checking for Valid Entries        
        #assert (type(player_id) is int)
        #assert (type(player_name) is str)   
        #assert (type(time_played) is float)
        #assert (type(player_pokemon) is dict)
        #assert (type(gyms_visited) is list)

        self.player_id = player_id
        self.player_name = player_name
        self.time_played = time_played
        self.player_pokemon = player_pokemon
        self.gyms_visited = gyms_visited
        
        print('Player object created >> ID: {0}, Name: {1}, Time Played: {2}, Pokemons collected: {3}, Gyms Visited: {4}'.format(player_id, player_name, time_played, player_pokemon, gyms_visited))

#Testcase with player1
player1 = Player(1,'Wilson',0.0,{}, [])

#Attribute check
print(player1.player_id)
print(player1.player_name)
print(player1.time_played)
print(player1.player_pokemon)
print(player1.gyms_visited)
````

## 7. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

Your function will:

1. Accept the `players` dictionary, `pokedex` dictionary, and a player_id as arguments.
2. For the specified player_id, look up that player's pokemon and their level(s).
3. Find and aggregate the attack and defense values for each of the player's pokemon from the `pokedex` dictionary.
4. Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
5. Return the player's power value.

Print out the pokemon power for each of your players.

````python
def calculate_power(players, pokedex, player_id):
    
    #Get the player_pokemon dict from the player based on the id
    pokemons_player_has = players[player_id].player_pokemon
    
    total_attack_pts = 0
    total_defense_pts = 0
    
    for k,v in pokemons_player_has.items():
        total_attack_pts += v['attack']
        total_defense_pts += v['defense']
    
    return total_attack_pts+total_defense_pts

#Individual print out
#print(calculate_power(players, pokedex, 1))
#print(calculate_power(players, pokedex, 2))

#Loop players list and print using the ID
for idx in players:
    print("{0}'s power is {1}.".format(players[idx].player_name, calculate_power(players, pokedex, idx)))
````

````python
## 8.1 ##
OuterList = []

for line in raw_pd.splitlines(): #Break into OuterList containing InnerList [Outer[Inner]]
    for i in [line]: 
        for j in [i.split(',')]:
            OuterList.append(j)

Cleaned_Pokedex = []

for InnerList in OuterList:
    InnerList2 = []
    for items in InnerList:
        if items == '': #Check for missing values and replace with -1
            items.replace('', -1) 
        items = items.strip('"')
        if not items.isdigit(): #Check to see if it can be converted to Float
            InnerList2.append(items) #Append to innerlist
        else:
            InnerList2.append(float(items))#Convert to Float and Append to innerlist
        
    Cleaned_Pokedex.append(InnerList2)

pprint(Cleaned_Pokedex)
````

Write a function that recreates the pokedex you made before, but with the data read in from the full pokemon file. The `PokedexNumber` should be used as the `pokemon_id` key values for the dictionary of pokemon.

````python
#function returns a dictionary
def find_pokemon(pokedex_raw, poke_id): #pokedex_raw = raw read-in from csv , poke_id = ID of the Pokemon to return
        
        cleaned_pokedex = [[l if not l.isdigit() else float(l) for l in k] for k in [[j.strip('"') for j in i] for i in [[line][0].split(',') for line in pokedex_raw.splitlines()]]]
        
        headers = cleaned_pokedex[0] #['PokedexNumber', 'Name', 'Type', 'Total', 'HP', 'Attack', 'Defense', 'SpecialAttack', 'SpecialDefense', 'Speed']
        
        selected_pokemon = [j for j in [i for i in cleaned_pokedex] if j[0] == poke_id] #Get pokemon(s) from the ID
        
        #print(selected_pokemon) #Print out checks to see what was found by the ID
        #print(len(selected_pokemon)) #check the len to see the amount of pokemon found with that ID as ID may not be unique
        
        master_list = []

        if len(selected_pokemon) > 1: #if the list of selected pokemon is larger than 1
            
            for pokemons in selected_pokemon:
                return_dict = {}
                for i in range(len(pokemons)):
        
                    return_dict[headers[i]] = pokemons[i]  
                
                master_list.append(return_dict)
                
            return master_list

        elif len(selected_pokemon) == 0:
            pprint("No Pokemon exist with that ID")
        
        else: #if the list of selected pokemon is 1
            
            return_dict = {}
            for i in range(len(selected_pokemon[0])):
                return_dict[headers[i]] = selected_pokemon[0][i]
            
            return return_dict
            
#Testing the function, ID=100 returns 1 pokemon, trying with ID = 6 returns more than 1 pokemon
find_pokemon(raw_pd, 100)
                
````
